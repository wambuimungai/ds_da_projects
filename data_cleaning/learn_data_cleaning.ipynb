```python
'''
Answer these questions before cleaning
What are the features?
What are the expected types (int, float, string, boolean)?
Is there obvious missing data (values that Pandas can detect)?
Is there other types of missing data that’s not so obvious (can’t easily detect with Pandas)? '''



import pandas as pd
import numpy as np

missing_values = ['n/a', 'na', '--']
df = pd.read_csv('property_data.csv', na_values = missing_values)

print(df.head())
```

What are the features and expected data type?

PID - Property ID (float/int)
ST_NUM - Street Number (float/int)
ST_NAME - Street Name(string)
OWN_OCCUPIED- Owner occupied residence(string)
NUM_BEDROOMS - Number of bedrooms(float/int)
NUM_BATH- Number of bathrooms(int/float)
SQ_FT- size in square feet(int)

```python
print(df)
print(df.isnull())
```

```python
cnt=0

for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
print(df.isnull().sum())
```

```python
print(df.isnull().values.any())
```

```python
print(df.isnull().sum().sum())
```

```python
#fill in the property IDs 
df['PID'] = df.PID.ffill() + df.groupby(df.PID.notnull().cumsum()).cumcount() * 1000 #multiply by 1000 to increment the 5th position of the ID
df
```

```python
median_stnum = df['ST_NUM'].median()
df['ST_NUM'].fillna(median_stnum, inplace=True)
df
```

```python
df['OWN_OCCUPIED'] = df['OWN_OCCUPIED'].ffill()
df
```

```python
max_bds = df['NUM_BEDROOMS'].max()
df['NUM_BEDROOMS'].fillna(max_bds, inplace=True)
df
```

```python
df.info()
```

```python
df['NUM_BATH'] = pd.to_numeric(df['NUM_BATH'], errors='coerce').ffill()
df
```

```python
median_sqft = df['SQ_FT'].median()
df['SQ_FT'].fillna(median_sqft, inplace=True)
df
```
